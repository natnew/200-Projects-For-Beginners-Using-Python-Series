```python
#Get data from Yahoo Finance 
#finance.yahoo.com
from yahoo_fin import stock_info
```

```python
#Tesla Inc
live_price = stock_info.get_live_price("TSLA")
print(round(live_price, 2), "UDA")
```

```python
#Get company info
stock_info.get_company_info("TSLA")
```

```python
#Get company officers
stock_info.get_company_officers("TSLA")
```

```python
#Get earnings history
stock_info.get_earnings_history("TSLA")
```

```python
#Get earnings financials
stock_info.get_financials("TSLA")
```

```python
#Get a quote table back as a dataframe
stock_info.get_quote_table("TSLA", dict_result = False)
```
